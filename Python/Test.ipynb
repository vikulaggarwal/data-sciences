```python
# find the longest consecutive sub-string in alphabetical order in a given string. 
# Note that the order determination is case insensitive, i.e. 
# the string "Ab" is considered to be in alphabetical order. 
# You can assume that the input will not have string where the number of possible consecutive sub-strings in alphabetical order is 0.
# i.e. the input will not have a string like "zxec".
string=input()
prevChar = ""
curr_longest = ""
longest = ""
orig = string
for char,origchar in zip(string.lower(),orig):
    if (prevChar<=char):
        curr_longest += origchar
        if len(curr_longest) > len(longest):
            longest= curr_longest
    else:
        curr_longest = origchar
    prevChar = char
print( longest )
```

```python
# if a palindrome can be created
string=input()
from collections import Counter 
res = Counter(string) 
print(res)
count=0
for key in res:
    if (res[key]%2!=0) & (count==1):
        count=2
        break
    if (res[key]%2!=0) & (count==0):
        count=1
if(count>1):
    print('No')
else:
    print('Yes')
```

```python
# The minion game
S='Australia'
Vowels = ('A' , 'a' , 'E' , 'e' , 'I' , 'i' , 'O' , 'o' , 'U' , 'u')

l = len(S)

sanskar = 0
mahima  = 0

for i in range (l):
    if S[i] in Vowels:
        mahima += l - i
    else:
        sanskar += l - i
        
if sanskar == mahima:
    print ("Draw")
elif sanskar > mahima:
    print ("Sanskar", sanskar)
else:
    print ("Mahima", mahima)
```

```python
scores = (9, 9, 18)

def tri_angle(x, y, z):
    if (x*x + y*y < z*z):
        print('Obtuse')
    elif (x*x + y*y == z*z):
        print('Right')
    elif (x*x + y*y > z*z):
        print('Acute')


x, y, z = sorted(scores)

if any([x >= y+z, y >= x+z, z >= x+y]):
    triangle_type = "Invalid"
elif all([x==y, x==z]):
    tri_angle(x,y,z)
    triangle_type = "Equilateral"
elif all([x!=z, z!=y, x!=y]):
    tri_angle(x,y,z)
    triangle_type = "Scalene"
elif any([
        all([x==z, x!=y]),
        all([x==y, x!=z]),
        all([z==y, z!=x]),
    ]):
    tri_angle(x,y,z)
    triangle_type = "Isosceles"
print(triangle_type)
```
