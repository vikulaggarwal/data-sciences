```python
df_group = df[(df.total_rural_urban == 'Total') & (df.age_group=='All ages')].groupby('area_name')
df_female_literacy = df_group['literate_females','total_females'].sum()
df_female_literacy['literacy_rate'] = df_female_literacy.literate_females/df_female_literacy.total_females
df_female_literacy.sort_values(by='literacy_rate',ascending=False)
```

```python
df_group = df[(df.total_rural_urban == 'Total') & (df.age_group=='All ages')].groupby('area_name')
df_literacy = df_group['literate_persons','total_persons'].sum()
df_literacy['literacy_rate'] = df_literacy.literate_persons/df_literacy.total_persons
df_literacy.sort_values(by='literacy_rate')
```
